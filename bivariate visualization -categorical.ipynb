```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
sns.countplot(data=df,x="survived",hue="class")
```

```python
x=pd.crosstab(df['survived'],df['class'])
```

```python
x
```

```python
sns.heatmap(x,annot=True,cmap="coolwarm")
```

```python
sns.barplot(x="class",y="age",hue="sex",data=df)
```
